```python
import pandas as pd
#import matplotlib as mpl
import matplotlib.pyplot as plt
```

Sample data.

```python
data = {'Classes':('Support', 'Assault', 'Engineer', 'Recon'),'Playtime':(15, 30, 45, 10)}
df = pd.DataFrame(data)
df
```

```python
df['Playtime'].plot(kind='pie',labels=df['Classes'],autopct='%1.f%%')
plt.title('Battlefield Class Preference')
plt.show()
```

```python
exp = (0,0,0,.1)
```

```python
plt.pie(df['Playtime'], explode = exp, labels=df['Classes'],autopct='%1.1f%%')
plt.title('Battlefield Class Preference')
plt.show()
```
