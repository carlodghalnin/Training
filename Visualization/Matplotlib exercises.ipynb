```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv')
df.head()
```

Exercise 1: Read Total profit of all months and show it using a line plot. Total profit data provided for each month. Generated line plot must include the following properties: 

X label name = Month Number

Y label name = Total profit

```python
plt.plot(df['month_number'],df['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.show()
```

Exercise 2: Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties: 

* Line Style dotted and Line-color should be red
* Show legend at the lower right location.
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker.
* Line marker color as read
* Line width should be 3

```python
plt.plot(df['month_number'],df['total_profit'],'o:r', label = 'Monthly Profit', linewidth = 3)
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.legend()
plt.show()
```

Exercise 3: Read all product sales data and show it using a multiline plot.
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product).

```python
products = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
units = df[products]

plt.xlabel('Month Number')
plt.ylabel('Units Sold')

for product in products:
    plt.plot(df['month_number'], df[product],'o-', label = product, linewidth = 3)
plt.legend(loc='upper left')
plt.show()
```

Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot.
Also, add a grid in the plot. gridline style should .

```python
plt.scatter('month_number', 'toothpaste', data = df)
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Units Sold')
plt.grid(linestyle='-')
plt.show()
```

Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

```python
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.bar('month_number', 'facecream', data = df, label = 'Face cream')
plt.bar('month_number', 'facewash', data = df, label = 'Face wash')
plt.legend()
plt.show()
```

Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk.

```python
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.bar('month_number', 'bathingsoap', data = df)
plt.title('Bath Soap Sales by Month')

plt.savefig('Bath Soap Sales')
plt.show()
```

Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges.

```python
plt.hist('total_profit', data=df)
plt.xlabel('Monthly Profit')
plt.ylabel('Count of Months')
plt.show()
```

Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart.

```python
products = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
units = []

for product in products:
    units.append(df[product].sum())
    
plt.pie(units, labels = products, autopct='%1.1f%%')
plt.show()
```

Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot.

```python
f, (ax1, ax2) = plt.subplots(2, sharex=True)

ax1.plot(df['month_number'],df['bathingsoap'])
ax1.set_ylabel('Units Sold')
ax1.set_title('Bath Soap Sales')

ax2.plot(df['month_number'],df['facewash'])
ax2.set_ylabel('Units Sold')
ax2.set_title('Face Wash Sales')

plt.xlabel('Month')
plt.show()
```

Exercise 10: Read all product sales data and show it using the stack plot.

```python
products = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']

plt.stackplot(df['month_number'], df[products].transpose(), data=df)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(products, loc='upper left')
plt.show()
```
