```python
from bs4 import BeautifulSoup
import requests
```

```python
with open('Test1.html') as html:
    soup = BeautifulSoup(html,'html.parser')
soup.prettify()
```

```python
#4. Write a Python program to extract all the URLs from the webpage
#python.org that are nested within 'li' tags from

#soup = BeautifulSoup(requests.get('https://www.python.org/').text, 'html5lib')

for pt in soup.find_all('li'):
    for link in pt.find_all('a'):
        print(link.get('href'))
```

```python
#5. Write a Python program to extract all the text from a given web page.

url = 'https://web.ics.purdue.edu/~gchopra/class/public/pages/webdesign/05_simple.html'
soup = BeautifulSoup(requests.get(url).text, 'html5lib')
    
for element in soup.find_all(True):
    print(element.text)
```

```python
#6. Write a Python program to retrieve the HTML code of the title, its text, and
#the HTML code of its parent.

url = 'https://web.ics.purdue.edu/~gchopra/class/public/pages/webdesign/05_simple.html'
#soup = BeautifulSoup(requests.get(url).text, 'html5lib')

print(soup.find_all('title'))
print(soup.find('title').text)
soup.find('title').parent
```

```python
#8. Write a Python program to remove the contents of a tag in a given html
#document.

url = 'https://web.ics.purdue.edu/~gchopra/class/public/pages/webdesign/05_simple.html'
soup = BeautifulSoup(requests.get(url).text, 'html5lib')

detagged = soup.basefont
print(detagged)
detagged.replace_with('')
print(soup.basefont)
```

```python
#9. Write a Python program to wrap an element in the specified tag and create
#the new wrapper.

url = 'https://web.ics.purdue.edu/~gchopra/class/public/pages/webdesign/05_simple.html'
soup = BeautifulSoup(requests.get(url).text, 'html5lib')

print(soup.title)
soup.title.string.wrap(soup.new_tag('b'))
print(soup.title)
```

```python
#10. Write a Python program to remove a tag from a given tree of html
#document and destroy it and its contents.

url = 'https://web.ics.purdue.edu/~gchopra/class/public/pages/webdesign/05_simple.html'
soup = BeautifulSoup(requests.get(url).text, 'html5lib')

detagged = soup.head
print(detagged)
detagged.replace_with('')
print('\n', soup.head)
print(soup.title)
```
